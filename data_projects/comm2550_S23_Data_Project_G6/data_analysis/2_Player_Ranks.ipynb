## First Step: Take the data from the CSV's and Order them to get ranks

For this section I am:
- Taking both season and pre all star game data, specifically PPG, APG, RPG
    - Splitting the data into All Star ranks against themselves and All Star Ranks against the field (top 100 NBA scorers)
        - making them ordered in descending fashion from top to bottom for each measurable stat
        - finding ranks from these newly ordered data

```python
import csv

with open('../data/player_stats_season.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    num_rows = sum(1 for row in reader)  # Count the number of rows
    header = reader.fieldnames  # Get the column names
    
num_cols = len(header) if header else 0  # Count the number of columns
    
print(f'Number of rows: {num_rows}')
print(f'Number of columns: {num_cols}')
print(f'Header: {header}')
```

```python
APG_filter = Full_season.sort_values('APG', ascending=False)
RPG_filter = Full_season.sort_values('RPG', ascending=False)
PPG_filter = Full_season.sort_values('PPG', ascending=False)
```

```python
Pre_APG_Filter = Pre_AS.sort_values('APG', ascending=False)
Pre_RPG_Filter = Pre_AS.sort_values('RPG', ascending=False)
Pre_PPG_Filter = Pre_AS.sort_values('PPG', ascending=False)
```

```python
filtered_data = list(Pre_AS[Pre_AS['AS'] > 0]['Name'])
```

```python
Pre_AS_LIST = Pre_AS[Pre_AS['Name'].isin(filtered_data)]
Pre_AS_PPG= Pre_AS_LIST.sort_values('PPG', ascending=False)
Pre_AS_APG= Pre_AS_LIST.sort_values('APG', ascending=False)
Pre_AS_RPG= Pre_AS_LIST.sort_values('RPG', ascending=False)
```

```python

Full_AS_LIST = Full_season[Full_season['Name'].isin(filtered_data)]
Full_AS_PPG = Full_AS_LIST.sort_values('PPG', ascending=False)
Full_AS_RPG = Full_AS_LIST.sort_values('RPG', ascending=False)
Full_AS_APG = Full_AS_LIST.sort_values('APG', ascending=False)
```

```python
Pre_AS_PPG['rank'] = pd.RangeIndex(start=1, stop=len(Pre_AS_PPG)+1)
Pre_AS_APG['rank'] = pd.RangeIndex(start=1, stop=len(Pre_AS_APG)+1)
Pre_AS_RPG['rank'] = pd.RangeIndex(start=1, stop=len(Pre_AS_RPG)+1)
```

```python
Full_AS_PPG['rank'] = pd.RangeIndex(start=1, stop=len(Full_AS_PPG)+1)
Full_AS_APG['rank'] = pd.RangeIndex(start=1, stop=len(Full_AS_APG)+1)
Full_AS_RPG['rank'] = pd.RangeIndex(start=1, stop=len(Full_AS_RPG)+1)
```

```python
APG_filter['rank'] = pd.RangeIndex(start=1, stop=len(APG_filter)+1)
RPG_filter['rank'] = pd.RangeIndex(start=1, stop=len(RPG_filter)+1)
PPG_filter['rank'] = pd.RangeIndex(start=1, stop=len(PPG_filter)+1)
```

```python
Pre_APG_Filter['rank'] = pd.RangeIndex(start=1, stop=len(Pre_APG_Filter)+1)
Pre_RPG_Filter['rank'] = pd.RangeIndex(start=1, stop=len(Pre_RPG_Filter)+1)
Pre_PPG_Filter['rank'] = pd.RangeIndex(start=1, stop=len(Pre_PPG_Filter)+1)
```

```python
Full_Season_AS_RanksAPG = APG_filter[APG_filter['Name'].isin(filtered_data)]
Full_Season_AS_RanksRPG = RPG_filter[RPG_filter['Name'].isin(filtered_data)]
Full_Season_AS_RanksPPG = PPG_filter[PPG_filter['Name'].isin(filtered_data)]

Pre_AS_RanksAPG = Pre_APG_Filter[Pre_APG_Filter['Name'].isin(filtered_data)]
Pre_AS_RanksRPG = Pre_RPG_Filter[Pre_RPG_Filter['Name'].isin(filtered_data)]
Pre_AS_RanksPPG = Pre_PPG_Filter[Pre_PPG_Filter['Name'].isin(filtered_data)]
```

### PPG Rank Difference

```python


# merge the two dataframes on the 'name' column
mergedPPG = pd.merge(PAS_PPG_Rank, FAS_PPG_Rank, on='Name')
```

```python
# subtract the data in the 'value' column of df2 from the data in the 'value' column of df1
mergedPPG['difference'] = mergedPPG['rank_x'] - mergedPPG['rank_y']
```

```python
mergedPPG.plot.bar(x='Name', y='difference')
plt.xlabel('All Stars')
plt.ylabel('Net Improvement')
plt.title('Net PPG Rank Improvements', loc='center', pad=20)
```

What I learn from the data:
- Some players drop or rise in their rank position depending on time 
- Some Players rose greatly like JJJ, Sabonis, and Jrue, but these raises didn't raise their status vs other All Stars 
- The Top players had the least amount of fluctuation
