## Set up 

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
Full_season= pd.read_csv('../data/player_stats_season.csv')
```

```python
Pre_AS= pd.read_csv('../data/player_stats_pre.csv')
```

```python
filtered_data = list(Pre_AS[Pre_AS['AS'] > 0]['Name'])
```

```python
Pre_AS_LIST = Pre_AS[Pre_AS['Name'].isin(filtered_data)]
Pre_AS_PPG= Pre_AS_LIST.sort_values('PPG', ascending=False)
Pre_AS_APG= Pre_AS_LIST.sort_values('APG', ascending=False)
Pre_AS_RPG= Pre_AS_LIST.sort_values('RPG', ascending=False)

Full_AS_LIST = Full_season[Full_season['Name'].isin(filtered_data)]
Full_AS_PPG = Full_AS_LIST.sort_values('PPG', ascending=False)
Full_AS_RPG = Full_AS_LIST.sort_values('RPG', ascending=False)
Full_AS_APG = Full_AS_LIST.sort_values('APG', ascending=False)
```

## Full Season Player Comparison Description ###

My question is "if we considered the full season's stats for certain players, would the outcomes of the All Star game selections be different?"

After reviewing my player rankings analysis, i can focus in on certain players who didn't make the all star 

- Trae Young (Huge Jump)

- Anthony Davis (Missed games)

- James Harden

My theory is that these players deserved to make the all star team if their full season stats were considered, and I will learn if this is true by comparing these players to the pre all star game stats of the chose all stars. 



## Trae Young ##

```python
# Add the row from df1 to df2
new_row = Full_season.iloc[12]

Pre_AS_LIST_Trae = Pre_AS_LIST.append(new_row, ignore_index=True)
```

```python
Pre_AS_LIST_Trae_PO = Pre_AS_LIST_Trae.sort_values('PPG', ascending=False)

Pre_AS_LIST_Trae_PO.plot.bar(x='Name', y='PPG')

bar_index = 13
bar_xpos = plt.gca().patches[bar_index].get_x()

plt.xlabel('All Stars + Trae')
plt.ylabel('PPG amount')
plt.title('Trae Young full season vs pre AS game All Stars PPG')
plt.legend()

# Highlight bar
highlight_color = "yellow"
highlight_alpha = 0.5
plt.axvspan(bar_xpos, bar_xpos+0.8, color=highlight_color, alpha=highlight_alpha)


```

Data Analysis:
- Trae ranks pretty high in the scoring ranks if he was placed in with other All Stars
- Trae beats out several other gaurds specifically
