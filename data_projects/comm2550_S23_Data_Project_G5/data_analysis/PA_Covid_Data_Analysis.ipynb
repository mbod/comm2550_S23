In this next section, we will be looking for the new case and death data 7 days before implementation of a new Level of social distancing, and track the number of cases in the 30 days following the date of implementation (explaining the 37 in the head). Then, we will graph the progression of cases over this span of 37 days in order to visualize the increase or decrease in cases.

```python

start_date = '5/22/20'
columns_to_select = ['new_cases', 'new_deaths']
rows_after_date1 = PA_df.loc[start_date:, columns_to_select].head(37)

print(rows_after_date1)
```

```python

rows_after_date1['new_cases'].plot(kind='line')

plt.title('New Cases During May 29, 2020 Yellow Phase Implementation in Pennsylvania')
plt.xlabel('Dates')
plt.ylabel('New Cases')

plt.show()
```
