### RQ3: What is the relationship between various weather measures and trends of Google search terms for dengue in Singapore?
#### H7: Precipitation is positively correlated with the number of dengue-related Google searches in Singapore.
#### H8: Humidity is positively correlated with the number of dengue-related Google searches in Singapore.
#### H9: Hours of sunshine is positively correlated with the number of dengue-related Google searches in Singapore.

_Note: Read the RQ1 and RQ2 notebook (in this order) before this notebook._

#### **Set up**

```python
import pandas as pd
import numpy as np
```

```python
google_trends = pd.read_csv('../data/cleaned/trends_clean.csv')
weather = pd.read_csv('../data/cleaned/weather_clean.csv')
```

```python
# read the RQ1 notebook to understand this step (separating the types of Google search terms)
google_trends['direct_terms'] = google_trends['dengue_fever'] + google_trends['dengue']
google_trends['indirect_terms'] = google_trends['mosquito_bite'] + google_trends['fever'] + google_trends['rashes'] + google_trends['rash'] + google_trends['mosquito']
```

```python
google_trends
```

```python
weather
```

```python
google_weather = google_trends.merge(weather, how='inner')
google_weather
```

```python
google_weather.columns
```

#### **H7: Precipitation is positively correlated with the number of dengue-related Google searches in Singapore.**

```python
# simplifying the dataframe for H7
h7_df = google_weather[['year','week','direct_terms','indirect_terms','rain_mm']]
h7_df
```

```python
corrs = []

for year, year_data in h7_df.groupby('year'):
    corrs.append(year_data.corr())
    
corrs_df = pd.concat(corrs)
corr_overall = corrs_df.groupby(corrs_df.index, sort=False).mean()
corr_overall.drop('year', inplace=True)
corr_overall.drop('year', axis=1, inplace=True)
corr_overall
```

```python
mask = np.zeros_like(corr_overall, dtype=bool)
mask[np.triu_indices_from(mask)] = True
corr_overall[mask] = np.nan
(corr_overall.style.background_gradient(cmap='twilight_shifted', axis=None, vmin=-1, vmax=1)
 .highlight_null(color='#f1f1f1').format(precision=3))
```

Precipitation and direct dengue-related terms have a weak and negative correlation. The same relationship is observed for precipitation and indirect dengue-related terms. Hence, we do not have evidence to confirm H7.

It is interesting that precipitation has a slight negative correlation with Google search trends, given that in RQ1, we saw a positive relationship between Google search trends and dengue cases and that in RQ2 the correlation between precipitation and dengue cases was so weak and also negative.

#### **H8: Humidity is positively correlated with the number of dengue-related Google searches in Singapore.**

```python
# simplifying the dataframe for H8
h8_df = google_weather[['year','week','direct_terms','indirect_terms','rh_mean']]
h8_df
```

```python
corrs2 = []

for year, year_data in h8_df.groupby('year'):
    corrs2.append(year_data.corr())
    
corrs2_df = pd.concat(corrs2)
corr2_overall = corrs2_df.groupby(corrs2_df.index, sort=False).mean()
corr2_overall.drop('year', inplace=True)
corr2_overall.drop('year', axis=1, inplace=True)
corr2_overall
```

```python
mask2 = np.zeros_like(corr2_overall, dtype=bool)
mask2[np.triu_indices_from(mask2)] = True
corr2_overall[mask2] = np.nan
(corr2_overall.style.background_gradient(cmap='twilight_shifted', axis=None, vmin=-1, vmax=1)
 .highlight_null(color='#f1f1f1').format(precision=3))
```

Humidity is weakly and negatively correlated to both direct dengue-related Google search terms and indirect dengue-related Google search terms. Hence, contrary to H8, it may be that humidity is negatively correlated with dengue searches.

It appears that for both direct and indirect terms, the correlation with humidity is negative, since we would expect that, in line with scientific research, higher humidity is related to greater reproduction of Aedes aegypti mosquitoes and thus higher transmission of dengue. If so, we would in turn expect that searches should go up to reflect the increase in dengue cases. Our finding to the contrary is therefore quite surprising.

```python
#simplifying the dataframe for H9
h9_df = google_weather[['year','week','direct_terms','indirect_terms','sun_hours']]
h9_df
```

```python
corrs3 = []

for year, year_data in h9_df.groupby('year'):
    corrs3.append(year_data.corr())
    
corrs3_df = pd.concat(corrs3)
corr3_overall = corrs3_df.groupby(corrs3_df.index, sort=False).mean()
corr3_overall.drop('year', inplace=True)
corr3_overall.drop('year', axis=1, inplace=True)
corr3_overall
```

```python
mask3 = np.zeros_like(corr3_overall, dtype=bool)
mask3[np.triu_indices_from(mask3)] = True
corr3_overall[mask3] = np.nan
(corr3_overall.style.background_gradient(cmap='twilight_shifted', axis=None, vmin=-1, vmax=1)
 .highlight_null(color='#f1f1f1').format(precision=3))
```

Hours of sunshine is weakly and positively correlated to direct and indirect dengue-related Google search terms. Hence, the correlations appear to support H9.

Overall, although we see positive correlations between hours of sunshine and Google search terms, these correlations are quite weak.

### Concluding comments

Overall, we find that many of our hypotheses were either rejected or even shown to be the contrary, at least based on the correlations and visual inspection of the line graph trends. One of the most intriguing is that neither rainfall nor relative humidity really correlates with dengue cases, which is opposite to what we would usually expect, given the increased breeding of Aedes aegypti mosquitoes in wetter environments. Nonetheless, mean temperature was still found to be one of the stronger predictors for dengue cases at lags of 0-1 week, in line with general expectations.

In terms of Google search trends, we find that search terms directly related to dengue are moderately and positively correlated with dengue cases with no lag intervals, which supports our overarching theory that people will tend to search more about dengue when they have dengue symptoms, which correspondingly reflects dengue case numbers. Perhaps, given the high awareness of dengue in Singapore, people with prolonged symptoms or symptoms characteristic of dengue tend to think of dengue and perform searches to confirm (or otherwise assure themselves) that they might have or do not have dengue.
