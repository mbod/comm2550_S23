# Background
* ## About the DOE
    * The New York City Department of Education (NYC DOE) is the largest school district in the United States
    * The DOE serves over 1.1 million students 
    * There are over 1,800 DOE schools across the five boroughs of New York City
    * The department often faces challenges due to underfunding and resource constraints
    <br/><br/>
    
    
    To address the challenges faced by the NYC DOE due to underfunding, a targeted approach to funding could be implemented. One potential solution could be to allocate funding based on the specific needs of each school, rather than providing a uniform amount of funding to all schools. This approach would require a thorough assessment of each school's needs, taking into account factors such as student body breakdowns and neighborhood demographics. By directing funding to schools with the greatest needs, the NYC DOE could ensure that resources are being allocated in a way that maximizes their impact. In addition, targeted funding could be used to support specific programs and initiatives, such as early childhood education or college and career readiness programs, that have been shown to have a positive impact on student outcomes.

### Recommendations

* ##### Target demographics by district, not by borough
    * As shown in the above geo analysis, borough data can be misleading because it encompasses so many districts. To target issues at their root, break down the city by school district for a more effective approach
    
    
* ##### Emphasize allocating free lunch, tutoring, extracurriculars, transportation
    * Increase funding for free and reduced-price lunch programs: To ensure that students have access to healthy and nutritious meals, the NYC DOE could allocate additional funding for free and reduced-price lunch programs. This could involve expanding the number of schools that offer free lunch programs, providing more funding to schools that already offer these programs, and improving the quality and nutritional value of school meals.
    * Provide additional funding for transportation: Transportation can be a significant barrier for low-income students who live far from their schools. The NYC DOE could allocate additional funding to support transportation programs, such as school buses, subway and bus passes, and other transportation options for students who need them.
    * Use data to allocate resources: By analyzing student demographics and school performance, the NYC DOE could determine which schools and students require additional support for free lunch and transportation programs. This data-driven approach could help ensure that resources are allocated where they are most needed.
    * Collaborate with community organizations: The NYC DOE could collaborate with community organizations to provide additional support for students who require free lunch and transportation. Community organizations can offer additional resources such as food pantries, food vouchers, and free transportation programs to support students and their families.
    
    


* ##### Targetted language support and assistive technology (especially Spanish and Mandarin) in Harlem and Flushing
    * Increase funding for English Language Learner (ELL) programs: ELL programs are essential to support students who are still learning English, and the NYC DOE could allocate more funding to these programs. This could include increasing the number of ELL teachers and support staff, providing additional professional development opportunities for teachers, and investing in resources such as instructional materials and technology.
    * Allocate funding based on poverty by neighborhood: The NYC DOE could target funding to schools located in neighborhoods with high poverty rates. These schools may have a higher concentration of low-income students who require additional support, such as after-school programs, social-emotional learning resources, and mental health services. Targeting funding to these schools could help ensure that all students have access to the resources and support they need to succeed.
    * Provide funding for community-based organizations: The NYC DOE could partner with community-based organizations to provide additional support for English language learners and low-income students. These organizations may have expertise in working with specific populations, and could provide after-school programs, mentorship opportunities, and other resources to support student learning and success.
    * Expand funding for early childhood education: Research shows that early childhood education is critical for student success, and the NYC DOE could allocate additional funding to support early childhood education programs. This could include expanding access to pre-kindergarten programs and providing professional development opportunities for early childhood educators.
