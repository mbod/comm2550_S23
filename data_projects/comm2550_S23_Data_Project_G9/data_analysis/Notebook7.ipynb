## County Statistics vs High, Low, and Medium Diabetes Prevalence

Using the data frames created in Notebook 3 I will focus on data frame subsets for high and low diabetes prevalence to determine if there are differences between counties in these groups that related to either income, education level, or insurance coverage.

After importing the data frames, I will make graphs for each factor and see how each factor interacts with the data.

### Set up data

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np

import warnings
warnings.simplefilter('ignore')
```

## Looking at Counties with Low Diabetes Prevalence Only

```python
#import low prevalence data frame made in Notebook 3
all_lowprev_county_stats_2017_df = pd.read_csv('../data/all_lowprev_county_stats_2017.csv')
all_lowprev_county_stats_2017_df
```

### Looking at % Uninsured

```python
all_lowprev_county_stats_2017_df.plot(kind='scatter', x='poverty_2017', y='uninsured_2017',c='db_prev',figsize=(12,10))
plt.xlabel('Poverty Rate (%)')
plt.ylabel('(%) Uninsured')
plt.title('Poverty vs. % Uninsured')
plt.subplots_adjust(bottom=0.5)
plt.show()
```

### Looking at income

#### Median household income

```python
all_lowprev_county_stats_2017_df.plot(kind='scatter', x='median_household_income_2017', y='uninsured_2017',c='db_prev',figsize=(12,10))
plt.xlabel('Median Household Income ($)')
plt.ylabel('(%) Uninsured')
plt.title('Median Household Income vs. % Uninsured')
plt.subplots_adjust(bottom=0.5)
plt.show()
```

#### Per capita income

```python
all_lowprev_county_stats_2017_df.plot(kind='scatter', x='per_capita_income_2017', y='uninsured_2017',c='db_prev',figsize=(12,10))
plt.xlabel('Per Capita Income ($)')
plt.ylabel('(%) Uninsured')
plt.title('Per Capita Income vs. % Uninsured')
plt.subplots_adjust(bottom=0.5)
plt.show()
```
