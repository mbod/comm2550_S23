## County Statistics vs Diabetes Prevalence (Education level)

Using the data frames created in Notebook 3 I will look for any correlations between diabetes prevalence and education level (high school and bachelors).

After importing the data frames, I will make graphs for each factor and see how prevalence level (very low, low, medium, high, and very high) and type of county (metropolitan and non-metropolitan) affect the data distribution.

### Set up data

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np

import warnings
warnings.simplefilter('ignore')
```

### Load and inspect the data

```python
#import the 2017 county data frame created in Notebook 3
all_county_stats_2017_df = pd.read_csv('../data/all_county_stats_2017.csv')
all_county_stats_2017_df
```

```python
#shape of data frame
num_rows, num_cols = all_county_stats_2017_df.shape

print(f'Data frame has {num_rows} rows and {num_cols} columns.')
```

## High School Education

```python
#Make a graph for high school gradutation rate vs. diabetes prevalence
all_county_stats_2017_df.plot(kind='scatter', x='hs_grad_2017', y='db_prev',figsize=(10,7))
plt.xlabel('High School Graduation Rate(%)')
plt.ylabel('Diabetes Prevalence(%)')
plt.title('Graduation Rate vs. Diabetes Prevalence by County')
plt.show()
```

```python
#Make a graph for high school gradutation rate vs. diabetes prevalence by county type
all_county_stats_2017_df.plot(kind='scatter', x='hs_grad_2017', y='db_prev',c='metro_colorcode',figsize=(10,7))
plt.xlabel('High School Graduation Rate(%)')
plt.ylabel('Diabetes Prevalence(%)')
plt.title('Graduation Rate vs. Diabetes Prevalence by County')
plt.show()
```

### Is there a correlation to poverty rate?

```python
#how does poverty rate correlate?
all_county_stats_2017_df.plot(kind='scatter', x='hs_grad_2017', y='db_prev',c='poverty_2017',figsize=(12,6))
plt.xlabel('High School Graduation Rate(%)')
plt.ylabel('Diabetes Prevalence(%)')
plt.title('Graduation Rate vs. Diabetes Prevalence by County')
plt.show
```

### Is there a correlation to % uninsured?

```python
#how does % uninsured correlate?
all_county_stats_2017_df.plot(kind='scatter', x='hs_grad_2017', y='uninsured_2017',c='db_prev',figsize=(12,7))
plt.xlabel('High School Graduation Rate (%)')
plt.ylabel('(%) Uninsured')
plt.title('Graduation Rate vs. % Uninsured')
plt.show()
```

```python
#how does % uninsured correlate?
#version 2, based on categorical prevalence level
all_county_stats_2017_df.plot(kind='scatter', x='hs_grad_2017', y='uninsured_2017',c='prev_colorcode',figsize=(10,8))
plt.xlabel('High School Graduation Rate (%)')
plt.ylabel('(%) Uninsured')
plt.title('Graduation Rate vs. % Uninsured')
plt.show()
```
